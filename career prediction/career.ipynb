```python
#importing the packages
import numpy as np
import pandas as pd
```

```python
#Reading the dataset
data = pd.read_csv("roo_data.csv")
data.head()
```

```python
data.columns
```

```python
#checking the dataset if any null values are present
data.info()
```

```python
#droping all null values
data.dropna(inplace =True)
```

# categorial values

```python
#extracting the categorial columns
categorial_cols =[cols for cols in data.columns if data[cols].dtype=='object']
categorial_cols
```

```python
Introvert = pd.get_dummies(data[["Introvert"]],drop_first=True)
Introvert.head()
```

```python
data['Suggested Job Role'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
suggested_job_role = encoder.fit_transform(data["Suggested Job Role"])
suggested_job_role = pd.DataFrame(suggested_job_role,columns =["Suggested Job Role"])
suggested_job_role.head()
```

```python
worked_In_teams = pd.get_dummies(data[['worked in teams ever?']],drop_first=True)
worked_In_teams.head()
```

```python
hard_smart_worker = pd.get_dummies(data[['hard/smart worker']],drop_first=True)
hard_smart_worker.head()
```

```python
salwork = pd.get_dummies(data[['Salary/work']],drop_first=True)
salwork.head()
```

```python
man_tech = pd.get_dummies(data[['Management or Technical']],drop_first=True)
man_tech.head()
```

```python
behaviour = pd.get_dummies(data[['Gentle or Tuff behaviour?']],drop_first=True)
behaviour.head()
```

```python
relation = pd.get_dummies(data[['In a Realtionship?']],drop_first=True)
relation.head()
```

```python
sal_expected = pd.get_dummies(data[['Salary Range Expected']],drop_first=True)
sal_expected.head()
```

```python
interest_books = pd.get_dummies(data[['Interested Type of Books']],drop_first=True)
interest_books.head()
```

```python
interest_games = pd.get_dummies(data[["interested in games"]],drop_first=True)
inputs_elder = pd.get_dummies(data[['Taken inputs from seniors or elders']],drop_first=True)
inputs_elder = pd.get_dummies(data[['Taken inputs from seniors or elders']],drop_first=True)
type_comp_settle = pd.get_dummies(data[['Type of company want to settle in?']],drop_first=True)
job_studies = pd.get_dummies(data[['Job/Higher Studies?']],drop_first=True)
interest_career = pd.get_dummies(data[['interested career area ']],drop_first=True)
interest_subjects = pd.get_dummies(data[['Interested subjects']],drop_first=True)
memory_score = pd.get_dummies(data[['memory capability score']],drop_first=True)
read_write = pd.get_dummies(data[['reading and writing skills']],drop_first=True)
olympiads = pd.get_dummies(data[['olympiads']],drop_first=True)
talent_test = pd.get_dummies(data[['talenttests taken?']],drop_first=True)
workshops = pd.get_dummies(data[['workshops']],drop_first=True)
certifications = pd.get_dummies(data[['certifications']],drop_first=True)
E_courses = pd.get_dummies(data[['Extra-courses did']],drop_first=True)
self_learning = pd.get_dummies(data[['self-learning capability?']],drop_first=True)
work_longer_time = pd.get_dummies(data[['can work long time before system?']],drop_first=True)
```

```python
new_data =pd.concat([data,work_longer_time,self_learning,E_courses,certifications,workshops,talent_test,olympiads,read_write
          ,memory_score,interest_subjects,interest_career,job_studies,type_comp_settle,inputs_elder,interest_games
          ,interest_books,sal_expected,relation,behaviour,man_tech,salwork,hard_smart_worker,worked_In_teams,Introvert],axis =1)
```

```python
self_learning.head()
```

```python
#removing the categorial columns in the data
new_data.drop(categorial_cols,inplace =True,axis =1)
```

```python
new_data.head()
```

```python
# we get the latest data which is numerical without the categorial values
new_data.head()
```

```python
#showing that there is no object values in the data
new_data.info()
```

```python
# deteriming the data
X = new_data
y = suggested_job_role
```

```python
X.shape
```

```python
y.shape
```

```python
#splitting the train data and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.info()
```

```python
#training the model using RandomForest
from sklearn.ensemble import RandomForestClassifier
model =  RandomForestClassifier(n_estimators=400)
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("the mean absolute error is ",mean_absolute_error(y_test,predictions))
```

```python
train_accuracy = model.score(X_train,y_train)
print("the accurcay of train data is",train_accuracy*100)
```

```python
test_accuracy = model.score(X_test,y_test)
print("the accurcay of test data is",test_accuracy*100)
```
